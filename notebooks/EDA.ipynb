```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("data/Gaming_and_Mental_Health.csv")
df.head()
```

```python
df1 = df[['age', 'gender', 'daily_gaming_hours', 'game_genre',
       'primary_game', 'gaming_platform', 'sleep_hours', 'sleep_quality',
       'sleep_disruption_frequency', 'face_to_face_social_hours_weekly', 'academic_work_performance']].copy()
df1.head()
```

```python
aux_freq = df1.groupby(["game_genre", "academic_work_performance"]).size().reset_index(name="frequencia")

fig = px.bar(
    aux_freq,
    x="game_genre",
    y="frequencia",
    color="academic_work_performance",        # segunda categoria vira cor/grupo
    barmode="group",            # "group" = lado a lado | "stack" = empilhado
    title="Frequência de Performance por tipo de jogo",
    labels={
        "game_genre": "Gênero do jogo",
        "frequencia": "Frequência",
        "academic_work_performance": "Performance na Faculdade e Trabalho"
    },
    text_auto=True
)

fig.show()
```

```python
fig = px.histogram(
    df1,
    x="age",
    color="academic_work_performance",
    barmode="group"
)
fig.show()
```

```python
fig = px.scatter(
    df1,
    x="daily_gaming_hours",
    y="sleep_hours",
    color="game_genre",
)

fig.show()
```
